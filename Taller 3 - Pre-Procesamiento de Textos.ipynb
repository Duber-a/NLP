# Punto 2:

- `[8pts]` Quitar palabras vacias
- `[9pts]` ¿Cuáles son las 10 palabras no vacias más usadas?

```python
## 2.1 Quitamos palabras vacias
import nltk   # Cargue de libreria para stopwords
nltk.download('stopwords') # Descargue lista de palabras vacias, descarga realizada una sola vez
from nltk.corpus import stopwords   
Lista_SW = stopwords.words('spanish')#Cargue de lista de palabras vacias 
```

```python
#Quitamos palabras vacias
Solitaria_Voz_SW = [palabra for palabra in Solitaria_Voz if palabra not in Lista_SW] #si la palabra no esta en la lista "stopwords" se dejara en el resultado else se quita
Solitaria_Voz_SW
```

```python
## 2.2 Conteo de palabras resultantes
conteo1 = []
for i in Solitaria_Voz_SW:            ## se identifica solo la cantidad
    conteo1.append(Solitaria_Voz_SW.count(i))       
    
Solitaria_Voz_SW_Cnt = sorted((list(zip(conteo1, Solitaria_Voz_SW))), reverse = True) #Se ordena y selecciona la palabra y cantidad
Solitaria_Voz_SW_Cnt  ## resultado toda la lista organizada
```

```python
2.2 ## se eliminan duplicados
Solitaria_Voz_SW_Cnt_final = []
for i in Solitaria_Voz_SW_Cnt:
    if i not in Solitaria_Voz_SW_Cnt_final:
        Solitaria_Voz_SW_Cnt_final.append(i)

Solitaria_Voz_SW_Cnt_final[0:10] #se seleccionan los primero 10
```
